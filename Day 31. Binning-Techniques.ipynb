```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/kaggle/input/competitions/titanic/train.csv')[['Age','Fare','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()

model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred=model.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print('Without Binning Model Accuracy',acc)
```

# Using Uniform


```python
discretizer = KBinsDiscretizer(
    n_bins=6,
    encode='ordinal',
    strategy='uniform'
)
```

```python
x_train_trans=discretizer.fit_transform(x_train)
x_test_trans=discretizer.transform(x_test)
```

```python
model1=DecisionTreeClassifier()
model1.fit(x_train_trans,y_train)

y_pred1=model1.predict(x_test_trans)

print('After Binning Accuracy (Uniform)',accuracy_score(y_test,y_pred1))
```

# Using Quantile

```python
discretizer1 = KBinsDiscretizer(
    n_bins=15,
    encode='ordinal',
    strategy='quantile'
)


x_train_trans1=discretizer1.fit_transform(x_train)
x_test_trans1=discretizer1.transform(x_test)

model2=DecisionTreeClassifier()
model2.fit(x_train_trans1,y_train)

y_pred2=model2.predict(x_test_trans1)

print('After Binning Accuracy (Quantile)',accuracy_score(y_test,y_pred2))
```

```python
discretizer3 = KBinsDiscretizer(
    n_bins=5,
    encode='ordinal',
    strategy='kmeans'
)


x_train_trans3=discretizer3.fit_transform(x_train)
x_test_trans3=discretizer3.transform(x_test)

model3=DecisionTreeClassifier()
model3.fit(x_train_trans3,y_train)

y_pred3=model3.predict(x_test_trans3)

print('After Binning Accuracy (K-Mean)',accuracy_score(y_test,y_pred3))
```
