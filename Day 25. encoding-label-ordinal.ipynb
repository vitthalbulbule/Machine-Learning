## Encoding Categorical Data | Ordinal Encoding | Label Encoder |

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Sex': ['Male', 'Female', 'Male', 'Female','Male', 'Female', 'Male', 'Female'],
    'Education': ['School', 'Graduate', 'Postgraduate', 'Graduate','School', 'Graduate', 'Postgraduate', 'Graduate'],
    'City': ['Pune', 'Mumbai', 'Delhi', 'Pune','Pune', 'Mumbai', 'Delhi', 'Pune'],
    'Salary': [20000, 30000, 40000, 35000,20000, 30000, 40000, 35000]
})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
X=df[['Sex','Education','City']]
y=df['Salary']

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
# Label Encoding

le=LabelEncoder()
x_train['Sex']=le.fit_transform(x_train['Sex'])
x_test['Sex']=le.transform(x_test['Sex'])
```

```python
x_train
```

```python
x_test
```

```python
# Orsinal Encoding
oe=OrdinalEncoder()


x_train[['Education']]=oe.fit_transform(x_train[['Education']])
x_test[['Education']]=oe.transform(x_test[['Education']])
```
