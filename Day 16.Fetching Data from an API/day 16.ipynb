```python
import requests
import time

base_url = "https://api.themoviedb.org/3/movie/changes"

params = {
    "api_key": "152828dad5c59a142ac80c434b3c1a43"   # regenerate your key
}

headers = {
    "User-Agent": "Mozilla/5.0"
}

all_data = [] 

for i in range(1, 55):
    params["page"] = i
    
    try:
        response = requests.get(base_url, params=params, headers=headers, timeout=10)
        
        if response.status_code == 200:
            data = response.json()
            
            # extract only 'results'
            results = data.get("results", [])
            all_data.extend(results)
         # avoid rate limit
    
    except Exception as e:
        print(f"Failed at page {i}: {e}")
        time.sleep(5)
```

```python
df = pd.DataFrame(all_data)

print(df.head())
print(df.shape)
```
