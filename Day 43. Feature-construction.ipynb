# Feature Construction

```python
df['family_size']=df['SibSp']+df['Parch']+1
```

```python
df.head()
```

```python
x=df[['Pclass','Age','family_size']]
y=df['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=10))
```

```python
def myfun(num):
    if num==1:
        return 0
    elif num>1 and num<=4:
        return 1
    else:
        return 2
```

```python
df['family_type']=df['family_size'].apply(myfun)
```

```python
df.head()
```

```python
df.drop(columns=['SibSp','Parch','family_size'])
```

```python
a=df[['Pclass','Age','family_type']]
b=df['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(a,b,test_size=0.2)
np.mean(cross_val_score(LogisticRegression(),a,b,scoring='accuracy',cv=20))
```
