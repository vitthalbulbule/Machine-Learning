```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get("https://www.ambitionbox.com/list-of-companies?page=1",headers=headers).text
```

```python

soup=BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div', class_='companyCardWrapper__primaryInformation')
```

```python
name=[]
rating=[]
rating_count=[]
about=[]

for i in company:
    name.append(i.find("h2").text.strip())
    rating.append(i.find_all('div',class_='rating_text')[0].text.strip())
    rating_count.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip())
    about.append(i.find('span',class_="companyCardWrapper__interLinking").text.strip())



d={'name':name,'rating':rating,'rating_count':rating_count,'about':about}
df=pd.DataFrame(d)
```

```python
final=pd.DataFrame()
for j in range (1,11):
    url="https://www.ambitionbox.com/list-of-companies?page={}".format(j)

    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text


    
    soup=BeautifulSoup(webpage,'lxml')

    company=soup.find_all('div', class_='companyCardWrapper__primaryInformation')
    
    name=[]
    rating=[]
    rating_count=[]
    about=[]
    
    for i in company:
        name.append(i.find("h2").text.strip())
        rating.append(i.find_all('div',class_='rating_text')[0].text.strip())
        rating_count.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip())
        about.append(i.find('span',class_="companyCardWrapper__interLinking").text.strip())
    
    
    
    d={'name':name,'rating':rating,'rating_count':rating_count,'about':about}
    df=pd.DataFrame(d)

    final=pd.concat([final,df],ignore_index=True)

    
```
