```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/kaggle/input/competitions/titanic/train.csv',usecols=['Age','Fare','Survived'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
# First Fill missing Values using Mean
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
si=SimpleImputer(strategy='mean')

x_train_tras=si.fit_transform(x_train)
x_test_trans=si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(x_train_tras,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=lr.predict(x_test_trans)
print('Accuracy Score',accuracy_score(y_test,y_pred))
```

# Using Missing Indictor

```python
s=SimpleImputer(add_indicator=True)

x_train_miss=s.fit_transform(x_train)
x_test_miss=s.transform(x_test)
```

```python
clf=LogisticRegression()
clf.fit(x_train_miss,y_train)
```

```python
pred=clf.predict(x_test_miss)
print('Accuracy_Score',accuracy_score(y_test,pred))
```

### Now see before Applying Missing Indictor Model Accuracy Was -> 0.69
### And after applying Missing Indictor Model Accuracy Is -> 0.70
