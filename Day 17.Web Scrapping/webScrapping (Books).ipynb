```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get("https://books.toscrape.com/catalogue/page-1.html",headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
Books=soup.find_all('li',class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
```

```python
name=[]
price=[]
availability=[]


for i in Books:
    name.append(i.find('h3').text.strip())
    price.append(i.find("p",class_='price_color').text.strip())
    availability.append(i.find('p',class_="instock availability").text.strip())
```

```python
d={'name':name,'price':price,'availability':availability}
df=pd.DataFrame(d)
```

```python
# Accessing Multiple Pages or All Pages

final=pd.DataFrame()

for j in range(1,100):
    url="https://books.toscrape.com/catalogue/page-{}.html".format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text

    soup=BeautifulSoup(webpage,'lxml')
    Books=soup.find_all('li',class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")

    
    name=[]
    price=[]
    availability=[]
    
    
    for i in Books:
        name.append(i.find('h3').text.strip())
        price.append(i.find("p",class_='price_color').text.strip())
        availability.append(i.find('p',class_="instock availability").text.strip())
    

    d={'name':name,'price':price,'availability':availability}
    df=pd.DataFrame(d)



    final = pd.concat([final,df],ignore_index=True)
```
