## One Hot Encoding With Top Categories

```python
# keep top 3 frequent categories
top_categories = df['Car_Model'].value_counts().index[:3]

# replace remaining with 'Other'
df['Car_Model'] = df['Car_Model'].apply(
    lambda x: x if x in top_categories else 'Other'
)

# one hot encoding
ohe = OneHotEncoder(sparse_output=False)

encoded = ohe.fit_transform(df[['Car_Model']])

encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out()
)

print(encoded_df)
```
