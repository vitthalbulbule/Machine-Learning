# Now I will Do the Separate pipeline for each feature and apply it using final Pipeline

```python
import pandas as pd
```

```python
df=pd.read_csv('/kaggle/input/competitions/titanic/train.csv')
```

```python
df.head()
```

```python
df=df.drop(columns=['PassengerId','Name','Cabin',"Ticket"],axis = 1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Age_pipeline -> For fill the missing values
# Embarked & Sex -> for filling missing values and one hot encoding
# column Transformer
# Final Pipeline
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
x=df.iloc[:,1:8]
y=df.iloc[:,0:1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

## Age_pipeline

```python

age_pipe=Pipeline([
    ('impute',SimpleImputer(strategy='mean'))
])
```

## cat_pipeline

```python
cat_pipe=Pipeline([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('OneHotEncode',OneHotEncoder(drop='first'))
])
```

## Column Transformer

```python
preprocessing = ColumnTransformer([
    ('age',age_pipe,[2]),
    ('category',cat_pipe,[1,6])
],remainder='passthrough')
```

# Final Pipeline

```python
from sklearn.linear_model import LogisticRegression

pipe=Pipeline([
    ('preprocess',preprocessing),
    ('Scaling',StandardScaler()),
    ('model',LogisticRegression())
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, pred)

print (acc)
```
