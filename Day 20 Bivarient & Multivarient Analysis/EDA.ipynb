```python
df['Family_size']=df['SibSp'] + df['Parch'] + 1
```

```python
df.head()
```

```python
# Create an family Type

#1 -> alone
# 2-4 -> Small
# >5 -> Large

def family_transform(num):

    if num==1:
        return 'alone'
    elif num>1 and num<5:
        return 'small'
    else:
        return 'large'
```

```python
df['Family_Type']=df['Family_size'].apply(family_transform)
```

```python
pd.crosstab(df['Survived'],df['Family_Type'],normalize='columns')*100
```
